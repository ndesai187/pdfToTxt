```python
import pdfplumber

myfile = open('output-test.csv', 'w')

with pdfplumber.open(r'Prudential-Standard-3PS-221-Aggregate-Risk-Exposures-(July-2017)_0.PDF') as pdf:
    for page in pdf.pages:
        text = page.extract_text()
        myfile.writelines(text)
        print(text)

myfile.close()
```

```python
from pdfminer.high_level import extract_pages
from pdfminer.layout import LAParams, LTTextBoxHorizontal, LTTextLineHorizontal

myfile2 = open('output-test-mine.txt', 'w')

path = r'Prudential-Standard-3PS-221-Aggregate-Risk-Exposures-(July-2017)_0.PDF'
for page in extract_pages(path):
    for element in page:
        if isinstance(element, LTTextLineHorizontal):
            print(element.get_text())
        if isinstance(element, LTTextBoxHorizontal):
            for line in element:
                print(line.get_text())
                myfile2.writelines(line.get_text())


myfile2.close()
```
